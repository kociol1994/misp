```python
zajecia = ["wf","informatyka","relax","mispy","piateczek"]

dnityg= ["Pon","Wto","Sro","Czwar","Pio"]

# wyprintuj zajecia od srody do piatku
print("  Zad 1")
for dzien, zajecie in zip(dnityg[2:], zajecia[2:]):
    print(dzien, ' ', zajecie)

print(' ')

# wyprintuj oprocz srody
for dzien, zajecie in zip(dnityg[:2], zajecia[:2]):
    print(dzien, ' ', zajecie)
for dzien, zajecie in zip(dnityg[3:], zajecia[3:]):
    print(dzien, ' ', zajecie)
```

# challenge 0.2

```python
zajecia = [["wf"],
           ["informatyka", "nauka czytania"],
           ["relax", "jedzenie"],
           ["mispy","mispy w domu"],
           ["piateczek", "praca domowa z mispow", "Ilovemispy", "mispy w czasie snu"]
          ]

dnityg= ["Pon","Wt","Sr","Czw","Pt"]

# zdefiniuj funkcje ktora poda ile jest zajec danego dnia
# lista_zajec_z_dnia -> ile ich jest
print("  Zad 2.1")
for dzien in enumerate(dnityg):
   zajete = dnityg[dzien[0]], len(zajecia[dzien[0]])
   print(zajete)
 
print(' ')
print("  Zad 2.2")
# posortuj dni po najbardziej zajetych
dni = [0, 1, 2, 3, 4]

for dzien in sorted(dni, key=lambda dzien: len(zajecia[dzien]), reverse=True):
    print(dnityg[dzien])

```

```python
print(' ')
print("  Zad 3")

# napisz program ktory zdajdzie dzien z najwieksza iloscia mispow
def czy_misp(day):
  tak = 0
  for i in zajecia[day]:
    if "misp" in i:
        tak+=1
  return tak

for dzien in sorted(dni, key=czy_misp, reverse=True):
    print(dnityg[dzien])
```
