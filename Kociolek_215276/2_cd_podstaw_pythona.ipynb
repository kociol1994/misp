```python
zajecia = [["wf"],
           ["informatyka", "nauka czytania"],
           ["relax", "jedzenie"],
           ["mispy","mispy w domu"],
           ["piateczek", "praca domowa z mispow", "Ilovemispy", "mispy w czasie snu"]
          ]

dnityg= ["Pon","Wto","Sro","Czwar","Pio"]

def dodaj_dwie_liczby(liczba1,liczba2):
    return liczba1+liczba2
```

```python
class DzienZajec():
    """obiekt trzymający dany dzień zajęć i przypisane mu metody"""
    
    
    def __init__(self, lista_na_dany_dzien, nazwa_dnia): #definicja
        self.lista_zajec= lista_na_dany_dzien
        self.dzien = nazwa_dnia
        
    # funkcja pozwalająca na sprawdzenie ilości występowania misp w danym dniu
    def ile_mispow(self):
        ile_m = sum([1 for zajecie in self.lista_zajec if "misp" in zajecie.lower()])
        return ile_m

    @property #zamienia na wlasciwosc   

    # Funkcja licząca ilość zajęć danego dnia
    def ile_zajec(self):
        return len(self.lista_zajec)
    
    # Wypisanie zajęć w danym dniu w postaci całego zdanie. Po wywołaniu klasy w funkcji print() zostanie wygenerowany właśnie ten tekst
    def __repr__(self):
        zajecia_do_stringa = ", ".join(self.lista_zajec)
        return "W dzien {} mam zajecia: {}".format(self.dzien, zajecia_do_stringa)
    
```

```python
# definicja nowej zmiennej z klasy DzienZajec
pio_zajecia = DzienZajec(["piateczek", "praca domowa z mispow", "Ilovemispy", "mispy w czasie snu"],
                        "piatek")
```

```python
# wywołanie funkcji liczącej ilosć zajęć 
pio_zajecia.ile_zajec
```

```python
# Tworzenie nowej klasy funkcji pozwalających na sprawdzenie numeru telefonu, określenie czy jest to numer komórkowy czy domowy 
# oraz sprawdzenie z jakiego miasta pochodzi ten numer w przypadku numeru domowego
# Klasa pozwala również wypisać poszczególne cyfry numeru słowami
class SprawdzNumer():

    def __init__(self, numer):
      self.anl = numer
      self.dlg = len(self.anl)
      if self.dlg == 9:
        print("Prawidlowy")
      else:
          print("Blad")

    def miasto(self):
      kierunkowy = int(self.anl[0])*10 + int(self.anl[1])
      if kierunkowy in Miasta.keys():
        return Miasta[kierunkowy]
      else:
        return "Inny numer"


    def __repr__(self):
      znaki = " "
      for i in range(self.dlg):
        nr = int(self.anl[i])
        znaki += str(Liczby[nr])
      return znaki
```

```python
kontrola = SprawdzNumer("426717280")
print(kontrola.miasto())
print(kontrola)
```
